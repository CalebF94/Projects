## Brief Exploratory Data Analysis

For classification, the target variable will be diagnosis where M=Malignant and B=Benign. For Regression the target variable will be  radius_mean. 

I'll begin with basic histograms to a get sense of the distribution for each variable. A distribution for diagnosis will not be included b/c it's binary

```python
y_class = bc['diagnosis']
bc=bc.drop(['diagnosis'], axis=1)
```

```python
fig, axs = plt.subplots(10, 3, figsize=(12, 40), sharey=True)

var_num=0
for col in range(0,3):
    for row in range(0,10):
        var_cur = bc.columns[var_num]
        sns.histplot(data=bc, x=var_cur, ax=axs[row, col], bins=10)
        var_num += 1
        axs[row,col].title.set_text(var_cur)
    

fig.tight_layout()
fig.savefig(plot_output + "\\histograms")
plt.show()
```
